```python
#NOT OPERATOR : for this loop, the result will always be the opposite boolean value of the statements collectively
x = 3
y = 2

if not((x==y) or (x+y == 6)):
  print("hello") #will be printed if the result of NOT operation is true
else :
  print('bye') #will be printed if the result is false
```

**SLICE OPERATOR**

```python
# slice operator slices the list or string between a specified start and stop index
fruits = ['apple','banana','orange','guava','kiwi']
text = 'this is sheldon cooper here'
print(fruits)
print('sliced list: ' ,fruits[1:3])
print(text)
print('sliced string: ',text[0:8])
```

**the format goes like this- name[start:stop]<br> the starting index will be included but the stop index wont. so , after slicing the string returned will have (stop-1)th index element.**<br>
**we can also either specify the start index or stop index**

```python
print(fruits[1::]) #here, apple starts with index 0 and banana with 1.
```

```python
print(text[:9])
```

**we can also use slice to insert elements into the list at a specified index**

```python
fruits = ['apple','banana','orange','guava','kiwi']
print(fruits)
#an element at the starting index
fruits[0:] = 'jackfruit'
print(fruits)
```

```python
#use .insert(index,'element') to insert element into the list

lists = ['apple','banana','orange','guava']
print('old: ',lists)
lists.insert(3,'kiiwii')
print('new: ',lists)
```

```python
#remove elements from list using del statement
print(lists)
del lists[0]
print(lists)
```

***del*  statement can remove items from any index. pop() method removes item only from the end**

```python
 # removing items by value using remove()
 fruits = ['apple','banana','orange','guava','kiwi']
 print(fruits)
 fruits.remove('kiwi')
 print(fruits)
```

```python
#sorting the list using sort()
cars = ['bmw', 'audi', 'toyota', 'subaru'] 
cars.sort()
print(cars)
```

```python
#reverse() - it just reverse the order of the list. it doesnt arrange in descending order.
cars.reverse()
print(cars)
```

```python
# A function can have multiple parameters
def func(x,text):
  print(x)
  if text == '1':
    print('text is 1')
  else:
    print('text is not 1')

func('hi','7')
```

```python
def age(age_value):
  if age_value>60:
    print('old')
  elif 30 < age_value <= 60:
    print('middle age')
  elif 19 < age_value <= 30:
    print('young')
  elif 13 < age_value <= 19:
    print('Teen')
  else:
    print('Kid')
  

age_value = int(input())
age(age_value)
```
